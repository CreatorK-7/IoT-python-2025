1. P511 배낭함수 문제의 1, 2에 적합한 코드를 골라서 작성하시오.

```python
def knapsack():
    # 테이블 초기화 (0으로 초기화된 2차원 리스트)
    array = [[0 for _ in range(maxWeight+1)] for _ in range(rowCount+1)]
 
    # maxWeight: 배낭의 최대 용량
    # weightrow: 각 물건의 무게
    # money: 각 물건의 가치

    # 동적 계획법 진행
    for row in range(1, rowCount + 1):
        for col in range(1, maxWeight+1):   
            if weightrow[row] > col: # 현재 물건의 무게가 배낭 용량을 초과하면
                array[row][col] = array[row-1][col] # 위의 값을 그대로 가져옴
            else:
                value1 = money[row] + array[row-1][col-weight[row]] # 현재 물건을 넣는 경우
                value2 = array[row-1][col] # 현재 물건을 안 넣는 경우
                array[row][col] = max(value1, value2) # 최대 가치 선택
    
    return array[rowCount][maxWeight] # 최적해 반환
```

2. P269 유명 맛집 대기줄 구현하기 문제를 while True:문을 사용하여 동적으로 입력을 받는 프로그램으로 변경작성하시오.

```python
## 함수 선언

# 큐가 가득 찼는지 확인
def isQueueFull():
    global SIZE, queue, front, rear 
    return rear == SIZE - 1  # rear가 큐의 마지막 인덱스에 도달하면 큐가 꽉 찬 상태

# 큐가 비었는지 확인
def isQueueEmpty(): 
    global SIZE, queue, front, rear
    return front == rear  # front와 rear가 같으면 큐가 비어 있는 상태

# 대기자를 큐에 추가
def enQueue(data): 
    global SIZE, queue, front, rear
    if isQueueFull():
        print('큐가 꽉 찼습니다.')  # 큐가 가득 찬 경우 추가 불가
        return
    rear += 1   # rear를 증가시켜 새로운 요소를 추가할 위치 지정
    queue[rear] = data  # 해당 위치에 데이터를 추가
    print(f'{data} 님이 대기 줄에 추가되었습니다.')

# 대기 줄에서 첫 번째 사람을 제거(식당에 입장시킴)
def deQueue():
    global SIZE, queue, front, rear
    if isQueueEmpty():
        print('큐가 비었습니다.')  # 큐가 비어 있으면 제거할 사람이 없다고 알림
        return None
    front += 1  # front를 증가시켜 첫 번째 요소 제거
    data = queue[front] # 제거된 데이터를 저장
    queue[front] = None # 제거된 위치를 None(빈 공간)으로 설정
    
    # 남아 있는 요소들을 한 칸씩 앞으로 이동
    for i in range(front + 1, rear + 1):
        queue[i - 1] = queue[i] # 각 데이터를 한 칸씩 앞당김
        queue[i] = None # 이동 후에 뒤쪽 자리는 None으로 설정
    front = -1  # front를 초기화시켜 큐가 비어 있는 상태로 만듦
    rear -= 1  # rear를 감소시켜 큐 크기 조정
    print(f'{data} 님이 식당에 들어갔습니다.')
    return data

# 큐에서 가장 앞에 있는 대기자 확인
def peek():
    global SIZE, queue, front, rear
    if isQueueEmpty():
        print('큐가 비었습니다.')  # 큐가 비어 있을 경우 확인할 요소 없음
        return None
    return queue[front + 1]  # 가장 앞에 있는 요소 반환

# 전역 변수 선언
SIZE = 5  # 큐의 크기 (최대 5명까지 대기 가능)
queue = [None for _ in range(SIZE)]  # 큐 초기화 (None 값으로 채움)
front = rear = -1  # front와 rear 초기화 (비어 있는 상태)

# 메인 코드
if __name__ == '__main__':
    # 프로그램이 실행    
    while True:
        print('\n=== 유명 맛집 대기줄 시스템 ===')
        print('1: 대기 추가 | 2: 입장 | 3: 대기 상태 확인 | 0: 종료')
        choice = input('선택: ')
        
        if choice == '1':
            name = input('대기자 이름 입력: ')
            enQueue(name)   # 대기자 추가
        elif choice == '2':
            deQueue()   # 큐에서 제거 (대기자 입장)
        elif choice == '3':
            print('현재 대기 줄 상태:', queue)  # 현재 대기 상태 출력
        elif choice == '0':
            print('프로그램을 종료합니다.') # 프로그램 종료
            break
        else:
            print('잘못된 입력입니다. 다시 선택하세요.')  # 잘못된 입력 처리
```

3. P309 편의점 판매된 물건 목록 출력하기 문제 역시 while True: 문으로 동적으로 입력받는 프로그램으로 변경하시오.
데이터의 종류는 [콘말차, 삿포로 맥주, 아카페라 벤티 헤이즐넛, 레어치즈푸딩, 척오리지널 메가 사워, 요아정 요거트컵, 페퍼로니피자주먹밥, 널담 슈톨렌, 딸기마시멜로케이크, 버니공쥬 딸기뚱카롱, 고추잡채호빵, 체다 슈크림붕어스낵] 

```python
import random

# TreeNode 클래스 정의: 이진 탐색 트리의 각 노드를 나타냄
class TreeNode():
    def __init__(self):
        self.left = None   # 왼쪽 자식 노드 초기화
        self.data = None   # 노드의 데이터 저장
        self.right = None  # 오른쪽 자식 노드 초기화

# 전역 변수
memory = []  # 트리에 삽입된 노드를 기록할 리스트
root = None  # 트리의 루트 노드
dataAry = ['콘말차', '삿포로 맥주', '아카페라 벤티 헤이즐넛', '레어치즈푸딩', '척오리지널 메가 사워', 
           '요아정 요거트컵', '페퍼로니피자주먹밥', '널담 슈톨렌', '딸기마시멜로케이크', '버니공쥬 딸기뚱카롱', 
           '고추잡채호빵', '체다 슈크림붕어스낵']  # 판매된 물건 목록

# 사용자가 동적으로 입력하는 물건 추가 (while True)
sellAry = []  # 사용자가 입력한 물건을 저장할 리스트
print('물건을 입력하세요. 종료하려면 "끝"을 입력하세요.')  # 사용자에게 입력 안내 메시지
while True:
    item = input('물건을 입력: ')  # 사용자로부터 물건을 입력받음
    if item == '끝':  # 사용자가 '끝'을 입력하면 입력 종료
        break
    elif item in dataAry:  # 입력받은 물건이 미리 정의된 목록에 있는지 확인
        sellAry.append(item)  # 물건이 목록에 있으면 sellAry에 추가
    else:
        print("잘못된 입력입니다. 목록에 있는 물건만 입력할 수 있습니다.")  # 그 외 잘못된 입력에 대한 안내

# 판매된 물건을 출력 (중복 포함)
print('오늘 판매된 물건(중복O) -->', sellAry)

# 이진 탐색 트리 구성
node = TreeNode()  # 첫 번째 물건에 대한 트리 노드 생성
node.data = sellAry[0]  # 첫 번째 물건의 데이터 삽입
root = node # 첫 번째 노드를 트리의 루트로 설정
memory.append(node) # 트리 노드를 메모리 리스트에 추가

# 나머지 판매된 물건들에 대해 트리 삽입
for name in sellAry[1:]:
    node = TreeNode()   # 새로운 노드 생성
    node.data = name    # 노드에 판매된 물건 데이터 삽입

    current = root  # 트리 탐색 시작
    while True:
        if name == current.data:    # 현재 노드와 삽입하려는 노드 데이터가 같으면 종료
            break
        if name < current.data: # 삽입할 노드가 현재 노드보다 작으면 
            if current.left == None:    # 왼쪽 자식이 없으면 
                current.left = node # 새 노드를 왼쪽 자식으로 삽입
                memory.append(node) # 새 노드를 메모리에 추가
                break  
            current = current.left  # 왼쪽 자식으로 이동
        else:   # 삽입할 노드가 현재 노드보다 크면 
            if current.right == None:   # 오른쪽 자식이 없으면 
                current.right = node    # 새 노드를 오른쪽 자식으로 삽입
                memory.append(node) # 새 노드를 메모리에 추가
                break
            current = current.right # 오른쪽 자식으로 이동

# 이진 탐색 트리 구성이 완료되었음을 알림
print('이진 탐색 트리 구성 완료!')

# 전위 순회 함수 (Pre-order Traversal)
def preoder(node):
    if node == None:  # 현재 노드가 None이면 종료
        return
    print(node.data, end=' | ')   # 현재 노드의 데이터를 출력
    preoder(node.left)  # 왼쪽 자식으로 이동하여 순회
    preoder(node.right) # 오른쪽 자식으로 이동하여 순회

# 판매된 물건 종류 출력 (중복 없이)
print('오늘 판매된 종류(중복X) -->', end=' | ')
preoder(root)  # 전위 순회를 통해 중복 없이 물건 출력
```

4. P473 Self Study 13-2 의 문제. 이진 검색 알고리즘을 풀면서 검색 횟수도 같이 출력하도록 구현하세요.

```python
import random

# 이진 검색 함수
def binSearch(ary, fData):
    pos = -1    # 초기 위치는 -1로 설정 (찾지 못했을 경우)
    start = 0   # 배열의 시작 인덱스
    end = len(ary) - 1  # 배열의 끝 인덱스
    search_count = 0    # 검색 횟수를 셈

    # start가 end보다 작거나 같은 동안 반복
    while start <= end: 
        search_count += 1  # 한 번 검색할 때마다 검색 횟수를 증가시킴
        mid = (start + end) // 2  # 배열의 중간 인덱스 계산
        
        # 찾고자 하는 값이 중간 값과 일치하면 해당 인덱스를 반환
        if fData == ary[mid]:
            return mid, search_count  # 검색된 인덱스와 검색 횟수 반환
        
        # 찾고자 하는 값이 중간 값보다 크면
        elif fData > ary[mid]:
            start = mid + 1  # start를 mid + 1로 변경하여 오른쪽 절반 탐색
        
        # 찾고자 하는 값이 중간 값보다 작으면 
        else:
            end = mid - 1  # end를 mid - 1로 변경하여 왼쪽 절반 탐색

    # 값이 배열에 없을 경우 초기위치와 검색 횟수를 반환
    return pos, search_count

# 랜덤한 숫자 100,000개 생성 (0 ~ 100000 사이)
dataAry = sorted([random.randint(0, 100000) for _ in range(100000)])  # 0부터 100000까지의 랜덤 숫자 100000개 생성 후 정렬
findData = random.randint(0, 100000)  # 찾을 값도 0부터 100000 사이에서 랜덤하게 선택

# 메인 코드 시작
# 배열의 일부 (처음 5개, 마지막 5개)만 출력하여 배열의 크기를 일부만 보여줌
print(f'배열 일부 --> {dataAry[:5]} ~~~ {dataAry[-5:]}')

# 이진 검색을 호출하여, 찾고자 하는 데이터와 검색 횟수를 반환받음
position, count = binSearch(dataAry, findData)

# 검색된 위치가 -1이면 값이 배열에 없다는 의미
if position == -1:
    print(f'{findData} (이)가 없네요.')
else:
    print(f'{findData} (은)는 {position} 위치에 있음')  # 찾은 데이터의 위치 출력

# 검색한 횟수를 출력
print(f"## {count}회 검색함")
```
